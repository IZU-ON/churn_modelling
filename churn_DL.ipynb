```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
labelencode = LabelEncoder()
df['Gender'] = labelencode.fit_transform(df['Gender'])
df.head()
```

```python
onehot = OneHotEncoder()
encoded = onehot.fit_transform(df[['Geography']]).toarray()
df_Geography = pd.DataFrame(encoded,columns=onehot.get_feature_names_out())
df_Geography.head()
```

```python
df = df.join(df_Geography.drop('Geography_Spain',axis=1))
df = df.drop('Geography',axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Exited',axis=1),df['Exited'],random_state=42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
scaler1 = StandardScaler()
x_train_scaled = scaler1.fit_transform(x_train)
x_train_scaled
```

```python
scaler2 = StandardScaler()
x_test_scaled = scaler1.fit_transform(x_test)
x_test_scaled
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential,layers
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8,activation='relu',input_dim=11))
model.add(Dense(16,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_scaled,y_train,epochs=100,batch_size=32,verbose=2)
model.evaluate(x_test_scaled,y_test,batch_size=32,verbose=2)
```

```python
model.get_weights()
```

```python
model1 = Sequential()
model1.add(Dense(8,activation='leaky_relu',input_dim=11))
model1.add(Dense(16,activation='leaky_relu'))
model1.add(Dense(32,activation='relu'))
model1.add(Dense(32,activation='leaky_relu'))
model1.add(Dense(32,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
model1.fit(x_train_scaled,y_train,epochs=100,batch_size=32,verbose=1)
model1.evaluate(x_test_scaled,y_test,batch_size=32,verbose=1)
```
